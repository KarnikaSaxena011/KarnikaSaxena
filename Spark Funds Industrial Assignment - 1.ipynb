# CHECKPOINT 5 :-Plots

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

##### A plot showing the fraction of total investments (globally) in venture, seed, and private equity, and the average amount of investment in each funding type. This chart should make it clear that a certain funding type (FT) is best suited for Spark Funds.

```python
funding_list=['venture','seed','private_equity']
p1=m1[m1.index.isin(funding_list) ]
p1_df=p1.groupby(by='funding_round_type')
p1_frac = pd.DataFrame(round(100*(p1['raised_amount_usd'])/sum(p1['raised_amount_usd']),2))
p1_avg=pd.DataFrame(p1_df['raised_amount_usd'].mean())
# p1_avg
# p1_frac
```

```python
sb.set(style='darkgrid')
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
s=sb.barplot(x=p1_frac.index,y=p1_frac['raised_amount_usd'],data=p1_frac)
s.set(xlabel='INVESTMENT TYPE', ylabel='PERC. of TOTAL RAISED AMOUNT USD',title='TOTAL INVESTMENT GLOBALLY')
plt.yscale('log')

plt.subplot(1,2,2)
s=sb.barplot(x=p1_avg.index,y=p1_avg['raised_amount_usd'],data=p1_avg)
s.set(xlabel='INVESTMENT TYPE', ylabel='AVERAGE RAISED AMOUNT USD',title='TOTAL INVESTMENT GLOBALLY')
plt.yscale('log')

plt.show()
```

##### A plot showing the top 9 countries against the total amount of investments of funding type FT. This should make the top 3 countries (Country 1, Country 2, and Country 3) very clear.

```python
sb.set(style='darkgrid')
plt.figure(figsize=(10,6))
s=sb.barplot(x=top9.index,y=top9['raised_amount_usd'],data=top9)
s.set(xlabel='TOP 9 COUNTRIES',ylabel='TOTAL AMOUNT OF INVESTMENT',title='TOTAL GLOBAL INVESTMENT COUNTRY-WISE')
plt.yscale('log')
plt.show()
```

##### A plot showing the number of investments in the top 3 sectors of the top 3 countries on one chart (for the chosen investment type FT).

```python
sb.set(style='darkgrid')
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
s=sb.countplot(x='main_sector',data=d1[d1['main_sector'].isin(d1_count['main_sector'])])
s.set(xlabel='TOP 3 SECTORS',ylabel='TOTAL NUMBER OF INVESTMENT',title='TOTAL NO. OF INVESTMENTS IN TOP 3 SECTORS OF TOP 3 COUNTRIES')
plt.yscale('log')
plt.xticks(rotation=90)

plt.subplot(1,3,2)
s=sb.countplot(x='main_sector',data=d2[d2['main_sector'].isin(d2_count['main_sector'])])
s.set(xlabel='TOP 3 SECTORS',ylabel='TOTAL NUMBER OF INVESTMENT',title='TOTAL NO. OF INVESTMENTS IN TOP 3 SECTORS OF TOP 3 COUNTRIES')
plt.yscale('log')
plt.xticks(rotation=90)

plt.subplot(1,3,3)
s=sb.countplot(x='main_sector',data=d3[d3['main_sector'].isin(d3_count['main_sector'])])
s.set(xlabel='TOP 3 SECTORS',ylabel='TOTAL NUMBER OF INVESTMENT',title='TOTAL NO. OF INVESTMENTS IN TOP 3 SECTORS OF TOP 3 COUNTRIES')
plt.yscale('log')
plt.xticks(rotation=90)
plt.show()
```
